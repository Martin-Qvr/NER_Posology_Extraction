## 2. Extracting useful data

```python
URL = "https://www.vidal.fr/medicaments/substances/liste"
principes_actifs = []
for i in range(26):
    letter = alphabet[i]
    aux = URL + "-" + letter + ".html"
    html_text = requests.get(aux).text
    soup = BeautifulSoup(html_text, 'html.parser')
    rows = soup.find_all('div', {"class": "list"})
    for row in rows:
        print(row.get_text())
        principes_actifs_aux = [text for text in row.stripped_strings]
        principes_actifs.extend(principes_actifs_aux)
```
